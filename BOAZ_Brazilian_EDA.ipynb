## 파생변수 

### 배송관련 파생 변수
- 가설1. 단순하게 주문일로부터 배송일이 빠르면 만족도가 높을 것임
> order_purchase_timestamp - order_delivered_customer_date

- 가설2. estimated date보다 실제 도착이 빠르면 만족도가 높을 것임
> order_estimated_delivery_date - order_delivered_customer_date	


## 가격 관련 파생변수
- 가설1.가격이 높으면 만족도 기준이 높지 않을까
